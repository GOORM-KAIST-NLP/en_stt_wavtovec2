```python
from tqdm import tqdm
#wandb.init(project="Audio2Text", entity="raghavadurs", id="asr_5")
wandb.init()
config = wandb.config
config.learning_rate = 0.01
```

```python
#define speech to text function'
#https://huggingface.co/tasks/automatic-speech-recognition
pipe = pipeline("automatic-speech-recognition", "facebook/wav2vec2-base-960h")
def asr_transcribe(audio_file):
  with open(audio_file, "rb") as f:
    data = f.read()
  result = pipe(data)["text"]

  return result
```

```python
gradio_asr = gr.Interface(
    fn=asr_transcribe,
    title="Speech To Text with Wav2Vec2 데모",
    description="오디오 클립을 넣으면, 번역해줘요!!",
    inputs=gr.inputs.Audio(label="오디오 파일 업로드", type="filepath"),
    outputs=gr.outputs.Textbox(label="자동번역"),
)
```
